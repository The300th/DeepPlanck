```python
import numpy as np
import DeepPlanck as DP

model = DP.BaseModel()

X,z,Mtrue = model.ReadFits()
```

```python
import matplotlib.pyplot as plt
plt.imshow(X)
plt.show()
```

The prediction will be the $log_{10}$ of the mass. The dimensions of X must be (batch,96,96,1) and z is an array which corresponds to the redshift values.

```python
model.PredictMass(X,[z]) 
```
